### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg  35?

```python
#your code here
number_mpg35above = df.loc[df.mpg>=35].shape[0]
print(number_mpg35above)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
origin_ofallcars=df.loc[(df.horsepower > 100) & (df.weight < 3000),'origin']
print(origin_ofallcars.value_counts().idxmax())
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
meanaxxforjapan=df.loc[df.origin == 'japan','acceleration'].mean()
print(round(meanaxxforjapan,2))
```

### 5. Which year had the highest average mpg?

```python
#your code here
avg_mpg_by_year = df.groupby('model_year')['mpg'].mean()
highest_avg_year = avg_mpg_by_year.idxmax()
print(highest_avg_year)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
median_mpg = df['mpg'].median()
filtered_df = df[df.mpg>median_mpg]
maxhptoweight = filtered_df.loc[(filtered_df.index == filtered_df.index.max())]
for name in maxhptoweight['name']:
    print(name)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python

avg_mpg_per_year = df.groupby(['origin', 'model_year'])['mpg'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.lineplot(data=avg_mpg_per_year, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
plt.figure(figsize=(20, 10))
sns.scatterplot(data=df,
                x='horsepower',
                y='weight',
                hue='origin',
                size='mpg',
                hue_order=['japan', 'europe', 'usa'],
                sizes=(df['mpg'].min()*4, df['mpg'].max()*5)) 

plt.title('Scatterplot of Horsepower vs Weight by Origin and MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.show()
```
